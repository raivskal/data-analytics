#3. Feature Scaling

Since Salaries vary from levels as per range -> we shall create feature scaling.

To do Feature Scaling, all vactors (here, *independant* and *dependant*) shall be 2D-arrays. 

We know, that *dependand* vector is 1D array - let's start with reshaping of it.

```python
dependent_var_reshaped = dependent_var.reshape(len(dependent_var),1)
dependent_var_reshaped
```

Now, we can do our feature scaling.

```python
from sklearn.preprocessing import StandardScaler

sc_X = StandardScaler()
sc_independent_var = sc_X.fit_transform(inpdependet_var)

sc_y = StandardScaler()
sc_dependent_var_reshaped = sc_y.fit_transform(dependent_var_reshaped)
```

```python
sc_independent_var #notika scalings
```

```python
sc_dependent_var_reshaped
```

# 5. Training of Machine Learning Model with Linear Regression

```python
from sklearn.svm import SVR

svr_reg_model = SVR(kernel="rbf") #iekavās nosaka kerneli

svr_reg_model.fit(sc_independent_var, sc_dependent_var_reshaped)
```

**Radial Basis Function** kernel - is one of most popular kernels in SVR.

# 6. Run a Prediction

```python
y_pred = svr_reg_model.predict(sc_X.transform([[7.5]])) #veic scaling transformcāciju
y_pred
```

```python
sc_y.inverse_transform(y_pred) #veic backwards scaling transformcāciju
```
