# Zip() Function

It works on e.g. two lists and merge them into 1 dictionary.

```python
BMW    = [1.1, 1.6, 1.8, 2.0, 2.5, 3.0, 4.0]
Subaru = [1.5, 1.6, 1.8, 2.0, 2.2, 2.5, 3.0]

list_of_engines = list(zip(BMW, Subaru))
list_of_engines
```

Next - pass zipped list into Pandas.

```python
import pandas as pd

df = pd.DataFrame(list_of_engines, columns = ["BMW", "Subaru"])
df
```

Can we have more columns to be zipped?

```python
Mercedes = [1.3, 1.5, 1.8, 2.0, 2.3, 2.5, 2.8]
```

```python
list_of_engines_2 = list(zip(BMW, Subaru, Mercedes))

df_2 = pd.DataFrame(list_of_engines_2, columns = ["BMW", "Subaru", "Mercedes"])
df_2
```

Faster option. :)

```python
df["Mercedes"] = Mercedes
df
```

# map(), apply(), applymap()

Let's summarize transformational functions we've learnt and check applymap() one.


*   apply() is used to apply a function along an axis of the DataFrame or on values of Series.
*   applymap() is used to apply a function to a ***DataFrame* elementwise**.
*   map() is used to substitute each value in a **Series** with another value.

# Parallel Processing

```python
#darba dalīšana daļās. Izmanto, kad tiešām ir jēga, jo aizņem laiku un jāsaprot, vai tā vērts
import numpy as np
from multiprocessing import Pool 

def parallel_processing(df, func):
  df_split = np.array_split(df, 2) #sadala 2. daļās
  pool = Pool(2) # 2threads
  df = pd.concat(pool.map(func, df_split))
  pool.close()
  pool.join()
  return df
```

```python
def add_feature(df):
  df["avg_open_close"] = np.mean([df.Open, df.Close], axis=0)
  return df
```

```python
%%timeit
result = parallel_processing(df_,add_feature)
```
