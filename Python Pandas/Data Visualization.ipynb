# Box-Plots and Percentiles

**Percentiles**

*Example 1:* You've got 75 percentile on an exam

Means that 75% of people passed exam **worser** than I did. But 25% of student passed exam **better** that I.

*Example 2:* Your height is at 80th percentile of entire Earth population.

That means, that, e.g. 1.85m tall person is taller than 80% of Earth population, but smaller than 20% of other people.

```python
students_1st_part = [33,38,43,45,48,49,50,52,55,61,61,64,65,66,67]
students_2nd_part = [68,69,70,71,72,73,77,77,78,79,79,80,84,84,91]
students = students_1st_part + students_2nd_part
students
```

**When you do work with percentiles - always make order in your data; like here - sort it!**

a.) Calculate 85th percentile ?

```python
len(students)*0.85
```

You always go for next number, thus it will be 26th number in a list

```python
students[25] #26.elements ir 25.pÄ“c kÄrtas listÄ
```

a.) Calculate 10th percentile ?

```python
len(students)*0.1
```

The same -> go for next number, thus it will be 4th number in a list


```python
students[3]
```

JÄpalasa par boxplotiem

Below you can see a box plot. It has quartiles:


*   Q2 - a median in a dataset
*   Q1 - a median of all items below Q2
*   Q3 - a median of all items above Q2
*   â€œmaximumâ€ - Q3 + 1.5*IQR
*   â€œminimumâ€ - Q1 - 1.5*IQR

Everythin below "min" and "max" is considered as **"outlier"**.



image.png

c.) Calculate first, second and third quartiles (Q1, Q2, Q3)

Q2 is most easier to find, because it is a median of the data.

```python
import numpy as np

q2 = np.median(students)
q2 #mediÄna Å†em skaitli pa vidu
```

Q1 is median of first half of a data-set.

```python
q1 = np.median(students_1st_part)
q1
```

Q3 is median of second half of a data-set.

```python
q3 = np.median(students_2nd_part)
q3
```

# Violin Plot

Violin plots are used to visualize data distributions, displaying the range, median, and distribution of the data.

A violin plot is a hybrid of a box plot and a kernel density plot, which shows peaks in the data.

```python
df_concat_clean
```

```python
plt.figure(figsize=(10, 8))
data_btc = df_concat_clean[df_concat_clean.Symbol == "BTC"]["Close"]
data_btc.reset_index(drop=True, inplace=True)

sns.violinplot(data=data_btc)
```

*   white dot represents the median
*   thick bar in center represents box from box-plot (an interquartile range)
*   thin line whiskers as in box-plot (outliers are behind min and max of whiskers)

Instead of Box-plot you can just show quartiles by having *inner="quartile"*

```python
plt.figure(figsize=(10, 8))
sns.violinplot(data=data_btc, inner="quartile")
```

*inner="stick"* - will bring you data-lines from data set

```python
plt.figure(figsize=(10, 8))
sns.violinplot(data=data_btc, inner="stick")
```

In our case - density representation isn't so informative, since we have quite a dispersion in ETH. Let's explore how can we add scatters to the plot.

```python
kde = sns.jointplot(x="Open", y="Close", data=ltc_eth, hue='Symbol', kind='kde')

kde.plot_joint(sns.scatterplot)


# delattr(kde, 'hue')
# kde.plot_joint(sns.regplot, data=ltc_eth[ltc_eth.Symbol == "ETH"], scatter=False, color='red')

kde.hue = None #Å¡is skatoties pÄ“c kÄ¼Å«ddu paziÅ†ojuma
kde.plot_joint(sns.regplot, data=ltc_eth[ltc_eth.Symbol == "ETH"], scatter=False, color='red') #apvieno 4 grafikus vienÄ, kde, joint plot, scatter plot, regression
```
