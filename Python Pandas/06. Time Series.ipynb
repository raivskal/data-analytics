```python
import pandas as pd
from datetime import datetime
```

Since it is Time Series topic - we shall work with time and dates. :)
Let's create such dataframe.

```python
df_ts = pd.date_range(start="03/06/2020", end="13/06/2020", freq="H") #uztaisa data series pa stundām
df_ts
```

We have data in ISO 8601 format: *yyyy-mm-dd hh:mm:ss*

```python
df_ts = pd.DataFrame(df_ts, columns=["datetime"])
df_ts
```

When we have a Data Frame - let's populate it with additional data

```python
import numpy as np

df_ts["calls"] = np.random.randint(0, 10, size=(len(df_ts)))
df_ts
```

To address time of any event or fact in table (e.g. number of calls done in each hour) we shall turn *datetime* column into index.

```python
df_ts = df_ts.set_index("datetime")
```

```python
df_ts.head()
```

We already know, that we can iterate via specific dates (since here, datetime is index).

```python
df_ts.loc["2020-03-06 04:00:00"]
```

```python
df_ts.loc["2020-03-06 00:00:00":"2020-03-06 04:00:00"]
```

Moreover we can do re-sampling.

```python
df_ts.resample("D").sum()
```

```python
df_ts.resample("D").mean()
```

```python
df_ts.resample("T").mean()
```

**Resampling frequencies**

- 'min', 'T' - minute
- ‘H’ - hour
- ‘D’ - day
- ‘B’ - business day
- ‘W’ - week
- ‘M’ - month
- ‘Q’ - quarter
- ‘A’ - year
