Combining logical expressions

```
wine_df.loc[(wine_df.country == "France") & (wine_df.price < 15.0)] #filtrēšana
```

```
wine_df.loc[(wine_df.country == "France") & (wine_df.price < 15.0) & (wine_df.winery.str.contains("Chateau|Château", case=False))] # | ir kā OR, & ir AND
```

```
wine_df.loc[(wine_df.country == "France") | (wine_df.price < 15.0) | (wine_df.winery.str.contains("Chateau|Château", case=False))] # | ir kā or
```

```
wine_df.loc[(wine_df.country == "France") | (wine_df.price < 15.0) | (wine_df.winery.str.contains("Chateau|Château", case=False)) & (wine_df.province == "Bordeaux")]
```

```
wine_df_bord = wine_df.loc[(wine_df.country == "France") | (wine_df.price < 15.0) | (wine_df.winery.str.contains("Chateau|Château", case=False)) & (wine_df.province == "Bordeaux")] #uztaisa jaunu dataframe
```

```
wine_df_bord.sort_values("points", ascending=False) #sortē pēc vīnu punktiem
```

```
wine_df_bord.sort_values(["points","price"], ascending=False) #sortē pēc vīnu punktiem un cenas
```

```
wine_df_bord.sort_values(["points","price"], ascending=[False, True]) #sortē pēc vīnu punktiem un cenas, dilstoši un augoši
```
